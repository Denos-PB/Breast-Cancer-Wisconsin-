```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import PowerTransformer,StandardScaler
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
df.head()
```

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(8, 6))
ax = sns.countplot(x = df['diagnosis'],data=df,order = df['diagnosis'].value_counts().index,palette='Set1',edgecolor = 'black')
plt.tight_layout()
plt.show()
```

```python
miss_value = df.isnull().sum()
miss_value
```

```python
numeric_cols = df.select_dtypes(include='number').columns

SKEW_THRESHOLD = 1.0
transformer = PowerTransformer(method='yeo-johnson', standardize=True)

for col in numeric_cols:
    if (df[col] < 0).any():
        continue
    skew = df[col].skew()
    if skew >= SKEW_THRESHOLD:
        X = df[[col]].values
        df[col] = transformer.fit_transform(X)
        skew_n = df[col].skew()
        print(f"Transforming col: {col} with skew {skew:.2f} and after yeo-johnson transformation a new skew is {skew_n:.2f}")
```

```python
categorical_cols = df.select_dtypes(include='object').columns
categorical_cols
```

```python
num_corr = df.corr(numeric_only=True)

mask = np.triu(num_corr)

plt.figure(figsize=(15, 15))
sns.heatmap (num_corr,
             cmap = 'coolwarm',
             center = 0,
             annot=True,
             fmt = ".2f",
             square=True,
             mask=mask,
             linewidths=0.5,
             cbar_kws={"shrink":0.8,"label":"Correlation Coef"},
             annot_kws={'size' : 8}
)
plt.tight_layout()
plt.show()
```
